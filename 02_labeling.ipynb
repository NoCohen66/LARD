# Labeled dataset generation

First create your export configuration :

- dataset_name is the name you want for output, labeled dataset
- output_directory is where the labeled dataset will be saved
- the "add_dataset_for_export" method allows to add a directory containing multiple earth studio acquisitions, rather than adding them one by one.

```python
config = ExportConfig()
config.dataset_name = "train_dataset"
config.output_directory = "data"
config.add_dataset_for_export(dataset_name="earth_studio_v1",
                              dataset_folder = "scenarios",
                             dataset_type = DatasetTypes.EARTH_STUDIO)
```

Then launch the export and you are done :

```python
export_datasets(config)
```

## Advanced usages

- Export from a yaml configuration :

```python
config = ExportConfig("params/export_test_dataset.yaml")
export_datasets(config)
```

- generate a single label file from a single earth studio scenario :

```python
yaml_config = Path("scenarios/DAAG_05_500/DAAG_05_500.yaml")
export_labels(yaml_config)
```

- Generate a dataset from multiple scenarios directory :

```python
config = ExportConfig()
config.dataset_name = "test_dataset"
config.output_directory = "data"
config.add_dataset_for_export(dataset_name="VBL_Real",
                              dataset_folder = "scenarios/test/VBL_Test_Real",
                             dataset_type = DatasetTypes.REAL)
config.add_dataset_for_export(dataset_name="VBL_Synth",
                              dataset_folder = "scenarios/test/VBL_Test_Synth",
                             dataset_type = DatasetTypes.EARTH_STUDIO)
config.add_dataset_for_export(dataset_name="VBL_Full_approach",
                              dataset_folder = "scenarios/test/VBL_Test_Synth_Full_Approach",
                             dataset_type = DatasetTypes.EARTH_STUDIO)
export_datasets(config)
```
